```python
symbols = "!@#$%^&*-_"

def humanLikePassGenerate():
    # варіанти людських паролів
    r = randrange(100)
    if r < 20:
        password = passWithDateOnlyGenerate()
    elif r < 40:
        password = PassWithWordsGenerate()
    elif r < 60:
        password = passWithNameGenerate()
    elif r < 80:
        password = passWithNumberOnlyGenerate()
    else:
        password = passWithKeyPatternGenerate()
    return replaceLettersWithNumbersInPass(password)

def PassWithWordsGenerate():
    # варіанти з словами
    password = ""
    word = words[randrange(len(words))]
    symbol = symbols[randrange(len(symbols))]
    number = f'{randrange(2022)}'
    flag = randrange(125)
    if flag < 25:
        password = word + symbol + number
    elif flag < 50:
        password = number + symbol + word
    elif flag < 75:
        password = number + word
    elif flag < 100:
        password =  word + number 
    else:
        password = word + words[randrange(len(words))]
        if len(password) < 10:
            password+= words[randrange(len(words))]
    password = changeLetterRegisterInPass(password)
    return password

def passWithNameGenerate():
    # варіанти з іменами
    name = nameWords[randrange(len(nameWords))]
    symbol = symbols[randrange(len(symbols))]
    number = f'{randrange(2022)}'
    flag = randrange(100)
    if flag < 25:
        password = name + symbol + number
    elif flag < 50:
        password = number + symbol + name
    elif flag < 75:
        password = number + name
    else:
        password =  name + number
    password = changeLetterRegisterInPass(password)
    return password

def passWithKeyPatternGenerate():
    # варіанти слів, чисел і клавіатурних комбінацій
    keyPattern = keyboardWords[randrange(len(keyboardWords))]
    word = words[randrange(len(words))]
    number = randrange(10000)
    if random() > 0.5:
        password = f'{word}{keyPattern}'
    else:
        password = f'{keyPattern}{number}'
    password = changeLetterRegisterInPass(password)
    return password

def passWithNumberOnlyGenerate():
    # варіанти чисел
    numbers = "0123456789"
    password = ""
    r = randrange(6, 11)
    for i in range(r):
        password += numbers[randrange(len(numbers))]
    return password

def passWithDateOnlyGenerate():
    # варіанти дат
    day = randrange(1,31)
    day = (f'{day}', f'{day:02}')[randrange(2)]
    month = randrange(1,13)
    month = (f'{month}', f'{month:02}')[randrange(2)]
    year = randrange(1921, 2021)
    year = (f'{year}', f'{year % 100:02}')[randrange(2)]
    r = randrange(6)
    if r == 0:
        password = f'{day}.{month}{year}'
    elif r == 1:
        password = f'{year}.{month}{day}'
    elif r == 2:
        password = f'{day}.{month}.{year}'
    elif r == 3:
        password = f'{year}.{month}.{day}'
    elif r == 4:
        password = f'{day}{month}{year}'
    else:
        password = f'{year}{month}{day}'
    return password

def replaceLettersWithNumbersInPass(password):
    # Заміна деяких букв цифрами
    replace = {
        'o': '0', 'e': '3', 's': '5', 'i': '1', 'g': '9',
        'O': '0', 'E': '3', 'S': '5', 'I': '1', 'B': '8', 'G': '6',
        'A': '4', '0': 'O', '5': 'S', '8': 'B', '9': 'g', '1': 'I', }
    result = ""
    for i in range(len(password)):
        if password[i] in replace and random() < 0.2:
            result += replace[password[i]]
        else:
            result += password[i]
    return result

def changeLetterRegisterInPass(password):
    # Заміна регістру букв
    r = randrange(4)
    if r == 0:
        result = password.upper()
    elif r == 1:
        result = password.upper()
    elif r == 2:
        result = ""
        for i in range(len(password)):
            if (random() < 0.5):
                if password[i].islower():
                    result += password[i].upper()
                else:
                    result += password[i].lower()
    elif r == 3:
        result = password.capitalize()
    return result

def passRandomGenerate():
    # варіанти випадкового паролю
    SOURCE = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789_!."
    password = ""
    length = randrange(5, 12)
    for i in range(length):
        password += SOURCE[randrange(len(SOURCE))]
    return password

def selectRandomPassFromTop(fromTop):
    # вибір випадкового паролю з списку
    return fromTop[randrange(len(fromTop))]
```
