```python
from pathlib import Path
import bcrypt
import secrets
import os
import hashlib
import base64
import json

def getSalt(salt_n=16, salt_var=1):
    if salt_var == 1:
        salt = os.urandom(salt_n)
    elif salt_var == 2:
        salt = secrets.token_bytes(salt_n)
    elif salt_var == 3:
        salt = bcrypt.gensalt()
    return salt

def hexToBytes(s):
    # Удаление прбельных символов и добавление 
    # 0 слева к фрагментам нечётной длины
    s = ''.join('0'[:len(f) % 2] + f for f in s.split())
    return bytes(int(s[i: i + 2], 16) for i in range(0, len(s), 2))

# salt = os.urandom(16)
# salt = secrets.token_bytes(16)
# salt = bcrypt.gensalt()

def bytesToHex(bs):
    return ''.join(f'{b:02X}' for b in bs)

def hashPasswordsSHA1(passwords, salt=b''):
    results = []
    for password in passwords:
        hashedSHA1 = hashlib.sha1(password.encode() + salt)
        results.append(bytesToHex(hashedSHA1.digest()))
    Path('outFiles/HashedSHA1.csv').write_text('\n'.join(results))

def hashPasswordsMD5(passwords, salt=b''):
    results = []
    for password in passwords:
        hashedMD5 = hashlib.md5(password.encode() + salt)
        results.append(bytesToHex(hashedMD5.digest()))
    Path('outFiles/HashedMD5.csv').write_text('\n'.join(results))

def hashPasswordsSHA256(passwords, salt=b''):
    results = []
    for password in passwords:
        hashedSHA256 = hashlib.sha256(password.encode() + salt)
        results.append(bytesToHex(hashedSHA256.digest()))
    Path('outFiles/HashedSHA256.csv').write_text('\n'.join(results))

def hashPasswordsMD5_SHA256(passwords, salt=b''):
    results = []
    for password in passwords:
        hashedMD5 = hashlib.md5(password.encode() + salt)
        hashedSHA256 = hashlib.sha256(password.encode() + salt)
        results.append(bytesToHex(hashedMD5.digest()) + ',' +
                       bytesToHex(hashedSHA256.digest()))
    Path('outFiles/HashedMD5SHA256.csv').write_text('\n'.join(results))

def hashs_create():
    passwords = Path('outFiles/Passwords.txt').read_text().splitlines()
    hashPasswordsSHA1(passwords)
    hashPasswordsMD5(passwords)
    hashPasswordsSHA256(passwords)
    hashPasswordsMD5_SHA256(passwords)
#    hashPasswordsBcrypt(passwords)

hashs_create()
```
